```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define A, d, and computed solution w_hat
A = np.array([[1, 0], [1, -1], [0, 1]])
d = np.array([-1, 2, 1])

# Compute least squares solution
w_hat = np.linalg.solve(A.T @ A, A.T @ d)

# Compute projection of d onto the column space of A
d_hat = A @ w_hat

# Compute residual vector
r = d - d_hat

# Define a grid for the plane spanned by columns of A
u = np.linspace(-2, 2, 10)
v = np.linspace(-2, 2, 10)
U, V = np.meshgrid(u, v)

# Compute corresponding points in 3D
X = A[0, 0] * U + A[0, 1] * V
Y = A[1, 0] * U + A[1, 1] * V
Z = A[2, 0] * U + A[2, 1] * V

# Create plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, color='gray', alpha=0.5, edgecolor='none')

# Plot vectors
ax.quiver(0, 0, 0, d[0], d[1], d[2], color='r', label='d (target vector)', linewidth=2)
ax.quiver(0, 0, 0, d_hat[0], d_hat[1], d_hat[2], color='b', label='d_hat (projection)', linewidth=2)
ax.quiver(d_hat[0], d_hat[1], d_hat[2], r[0], r[1], r[2], color='g', label='Residual', linewidth=2)

# Labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Least Squares Projection in R3')
ax.legend()

plt.show()
```
