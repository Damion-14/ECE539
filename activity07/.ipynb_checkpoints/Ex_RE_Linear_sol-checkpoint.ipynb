```python
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# makes printing more human-friendly
np.set_printoptions(precision=3, suppress=True)
```

```python
data1 = {
    'x': np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])[:, None],
    'y': np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
}

data2 = {
    'x': np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])[:, None],
    'y': np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
}

data3 = {
    'x': np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])[:, None],
    'y': np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
}

data4 = {
    'x': np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])[:, None],
    'y': np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])
}
```

```python
# 1a) Fit the data
reg1 = LinearRegression().fit(data1['x'], data1['y'])
print('coef:', reg1.coef_)
print('intercept:', reg1.intercept_)
```

```python
# 1b) Predict and measure R2 scores
yhat1 = reg1.predict(data1['x'])
rsqr1 = r2_score(data1['y'], yhat1)
w1, b1 = reg1.coef_, reg1.intercept_
print(f'Model 1: y = {w1[0]:.2f}x + {b1:.2f} \t R2 = {rsqr1:.2f}')
```

```python
# 1c) Plot the data and predictions
x_range = np.linspace(0, 20, 100)[:, None]
plt.plot(data1['x'], data1['y'], 'r.', ms=15);
plt.plot(x_range, reg1.predict(x_range), 'r-');
```

```python
# 1d) Compute the linear regression for other datasets
reg2 = LinearRegression().fit(data2['x'], data2['y'])
yhat2 = reg2.predict(data2['x'])
rsqr2 = r2_score(data2['y'], yhat2)
w2, b2 = reg2.coef_, reg2.intercept_
print(f'Model 2: y = {w2[0]:.2f}x + {b2:.2f} \t R2 = {rsqr2:.2f}')

reg3 = LinearRegression().fit(data3['x'], data3['y'])
yhat3 = reg3.predict(data3['x'])
rsqr3 = r2_score(data3['y'], yhat3)
w3, b3 = reg3.coef_, reg3.intercept_
print(f'Model 3: y = {w3[0]:.2f}x + {b3:.2f} \t R2 = {rsqr3:.2f}')

reg4 = LinearRegression().fit(data4['x'], data4['y'])
yhat4 = reg4.predict(data4['x'])
rsqr4 = r2_score(data4['y'], yhat4)
w4, b4 = reg4.coef_, reg4.intercept_
print(f'Model 4: y = {w4[0]:.2f}x + {b4:.2f} \t R2 = {rsqr4:.2f}')
```

```python
# e) Plot all datasets and linear regression line
x_range = np.linspace(0, 20, 100)[:, None]

f, ax = plt.subplots(1, 4, figsize=(16, 3))
ax[0].plot(data1['x'], data1['y'], 'r.', ms=15)
ax[0].plot(x_range, reg1.predict(x_range), 'r-')
ax[1].plot(data2['x'], data2['y'], 'b.', ms=15)
ax[1].plot(x_range, reg2.predict(x_range), 'b-')
ax[2].plot(data3['x'], data3['y'], 'g.', ms=15)
ax[2].plot(x_range, reg3.predict(x_range), 'g-')
ax[3].plot(data4['x'], data4['y'], 'k.', ms=15)
ax[3].plot(x_range, reg4.predict(x_range), 'k-')
for axes in ax:
    axes.set_xlim(0, 20);
    axes.set_ylim(0, 15);
```

# Discuss observations

The R-square values for linear regression predictions remain consistent across various datasets. However, it's important to note that not all datasets are inherently suited for a linear regression model. For instance, datasets 2, 3, and 4 exhibit non-linear data distributions. Without visual aids, it becomes difficult to ascertain the aptness of a linear model for any given dataset, especially when they share similar performance metrics, such as the R2 value. Specifically, for dataset 2, a polynomial (quadratic) regression model would be a more fitting choice.
