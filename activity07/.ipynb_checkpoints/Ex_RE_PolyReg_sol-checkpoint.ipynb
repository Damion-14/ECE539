```python
import numpy as np
import matplotlib.pyplot as plt

# makes printing more human-friendly
np.set_printoptions(precision=3, suppress=True)
```

```python
# Load the data
colab=False
if colab:
    from google.colab import drive
    drive.mount('/content/drive')
    with open('/content/drive/539/data/Ex_poly_data.csv', 'r') as f: 
        data = np.genfromtxt(f,delimiter=',')
else:
    with open('Ex_poly_data.csv', 'r', encoding='utf-8') as f: 
        data = np.genfromtxt(f,delimiter=',')

plt.plot(data[:, 0], data[:, 1], '.', label='y [noisy observations]');
plt.plot(data[:, 0], data[:, 2], '-', label='d(x) [truth]');
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
data_trainval, data_test = train_test_split(data, train_size=0.85)
data_train, data_val = train_test_split(data_trainval, train_size=0.7/0.85)

x_train = data_train[:,0]
y_train = data_train[:,1]
y_true_train = data_train[:,2]

x_val = data_val[:,0]
y_val = data_val[:,1]
y_true_val = data_val[:,2]

x_test = data_test[:,0]
y_test = data_test[:,1]
y_true_test = data_test[:,2]
```

```python
from sklearn.linear_model import LinearRegression
def train_polyreg_model(x, y, order=3):
    X = np.stack([x ** (m+1) for m in range(order)], axis=1)
    # print('X:', X.shape, X[1]) # X should be an Nxm matrix, where each row is [x, x^2, ..., x^m]
    model = LinearRegression().fit(X, y)
    # print(model)
    return model
    
def predict_polyreg_model(model, x):
    order = len(model.coef_)
    X = np.stack([x ** (m+1) for m in range(order)], axis=1)
    return model.predict(X)

from sklearn.metrics import r2_score
order_list = range(1, 11)
r2_list = []
for order in order_list:
    model = train_polyreg_model(x_train, y_train, order=order)
    yhat_val = predict_polyreg_model(model, x_val)
    r2 = r2_score(y_val, yhat_val)
    r2_list.append(r2)

print('R2 Scores:', np.array(r2_list))

plt.plot(order_list, r2_list);
plt.xlabel('Polynomial Order (m)')
plt.ylabel('R2 Score')
plt.grid(True)
plt.show()
```

```python
optimal_order = 2
model = train_polyreg_model(x_train, y_train, order=optimal_order)
yhat_test = predict_polyreg_model(model, x_test)
r2 = r2_score(y_test, yhat_test)
print('R2 score: ', r2)

x_range = np.linspace(-0.5, 0.5, 100)
yhat_range = predict_polyreg_model(model, x_range)

plt.plot(data[:, 0], data[:, 1], '.', label='y [noisy observations]');
plt.plot(data[:, 0], data[:, 2], '-', label='d(x) [truth]');
plt.plot(x_range, yhat_range, '-', label='y predicted');
plt.legend()
plt.show()
```
